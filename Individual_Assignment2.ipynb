### (1) Methods and Plan

As stated in the previous assignment, the question we want to investigate can be focused on both inference and prediction. We have decided to shift our focus to prediction, as we believe it's more meaningful to predict the probability of being placed. So I'll use logistic regression to fit a model optimized for predicting a student’s likelihood of being placed based on given academic and extracurricular profile. All variables will be included in this model except StudentID which we have already omitted in the previous assignment and I plan to apply model selection in the later process. With the dataset split into training (70%) and testing (30%) sets, I'll also evaluate predictive performance of this model. This method is appropriate because PlacementStatus is a binary response with two levels: Placed and NotPlaced, and logistic regression models the probability of a categorical response and splitting dataset ensures that the model can generalize to unseen data. 

Logistic regression assumes the independence of observations, no extreme multicollinearity among predictors, linearity in the log-odds and a sufficient large sample size. 

Potential limitations include sensitivity to outliers and overfitting risks if predictors are correlated, which can be addressed by removing outliers and assessing variance inflation factors (VIF).

### (2)  Computational Code and Output

```r
set.seed(1234) # Make sure the code is reproducible

# Remove StudentID and split the dataset into training set and testing set
placement_split <- placement_data %>%
    select(-StudentID) %>%
    initial_split(prop = 0.7, strata = PlacementStatus)

placement_training <- training(placement_split)
placement_testing <- testing(placement_split)

# Full logistic regression model
logit_model <- glm(PlacementStatus ~ . ,
                   data = placement_training, family = binomial)
```

```r
# Check for multicollinearity
vif_values <- vif(logit_model)
vif_values
```

```r
# Make prediction on test dataset
prediciton <- placement_testing %>%
    mutate(pred = round(predict(logit_model, placement_testing, type = "response"))) %>%
    mutate(pred = if_else(pred == 0, "NotPlaced", "Placed"))
```

```r
# Check the performance of the model
placement_confusion_matrix <- 
    confusionMatrix(
    data = as.factor(prediciton$pred),
    reference = as.factor(prediciton$PlacementStatus),
    positive = "Placed"
)
placement_confusion_matrix
```
