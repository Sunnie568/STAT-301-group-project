### (1) Methods and Plan

As stated in the previous assignment, the question we want to investigate can be focused on both inference and prediction. We have decided to shift our focus to prediction, as we believe it's more meaningful to predict the probability of being placed. So I'll use logistic regression to fit a model optimized for predicting a student’s likelihood of being placed based on given academic and extracurricular profile. All variables will be included in this model except StudentID which we have already omitted in the previous assignment and I plan to apply model selection in the later process. With the dataset split into training (70%) and testing (30%) sets, I'll also evaluate predictive performance of this model. This method is appropriate because PlacementStatus is a binary response with two levels: Placed and NotPlaced, and logistic regression models the probability of a categorical response and splitting dataset ensures that the model can generalize to unseen data. 

Logistic regression assumes the independence of observations, no extreme multicollinearity among predictors, linearity in the log-odds and a sufficient large sample size. 

Potential limitations include sensitivity to outliers and overfitting risks if predictors are correlated, which can be addressed by removing outliers and assessing variance inflation factors (VIF).

### (2)  Computational Code and Output

```r
set.seed(1234) # Make sure the code is reproducible

# Remove StudentID and split the dataset into training set and testing set
placement_split <- placement_data %>%
    select(-StudentID) %>%
    initial_split(prop = 0.7, strata = PlacementStatus)

placement_training <- training(placement_split)
placement_testing <- testing(placement_split)

# Full logistic regression model
logit_model <- glm(PlacementStatus ~ . ,
                   data = placement_training, family = binomial)
```

```r
# Check for multicollinearity
vif_values <- vif(logit_model)
vif_values
```

```r
# Make prediction on test dataset
prediciton <- placement_testing %>%
    mutate(pred = round(predict(logit_model, placement_testing, type = "response"))) %>%
    mutate(pred = if_else(pred == 0, "NotPlaced", "Placed"))
```

```r
# Check the performance of the model
placement_confusion_matrix <- 
    confusionMatrix(
    data = as.factor(prediciton$pred),
    reference = as.factor(prediciton$PlacementStatus),
    positive = "Placed"
)
```

#### Results

```r
placement_confusion_matrix
```

The model predicted the likelihood of placement with a test accuracy of 80.44%, which demonstrates strong predictive performance on unseen data. Sensitivity (77.38%) and specificity (82.65%) showed that both "Placed” and “NotPlaced” students were effectively distinguished. The VIF values are all below 1.5 which confirms no multicollinearity issues. The full model containing 10 covariates performs well, but has the potential risk of overfitting, which requires optimization of model selection in the further process.

### Model Selection

We already prepared our training data with `model_matrix_X_train` and `matrix_Y_train`. We need to find the value of $\lambda$ that provides the largest average AUC when a LASSO penalty is used.

```r
model_matrix_X_train <- model.matrix(PlacementStatus ~ ., data = placement_training)[,-1]
head(model_matrix_X_train)

matrix_Y_train <- 
    as.matrix(placement_training$PlacementStatus, ncol = 1)
head(matrix_Y_train)
```

Since Ridge has been proposed as a method to address multicollinearity problems and we don't have this problem here, we prefer to use LASSO to select strong predictive models.

```r
set.seed(1234)
placement_cv_lambda_lasso <- 
  cv.glmnet(
       x = model_matrix_X_train, 
       y = matrix_Y_train,
       alpha = 1,
       family = "binomial",
       type.measure = "auc",
       nfolds = 10)
placement_cv_lambda_lasso
```

We can use the function `plot()` to visualize the CV-AUC values for each value of $\lambda$ in the grid.

```r
plot(placement_cv_lambda_lasso, 
     main = "Cross-Validation with Lasso Regression\n\n")
```

```r
lambda_max_AUC_lasso <- round(placement_cv_lambda_lasso$lambda.min, 4)
lambda_max_AUC_lasso

lambda_1se_AUC_LASSO <- round(placement_cv_lambda_lasso$lambda.1se, 4)
lambda_1se_AUC_LASSO
```

```r
plot(placement_cv_lambda_lasso$glmnet.fit, "lambda") +
    abline(v = log(lambda_max_AUC_lasso), col = "red", lwd = 3, lty = 2) +
    abline(v = log(lambda_1se_AUC_LASSO), col = "blue", lwd = 3, lty = 2)
```

```r
lambda_max_AUC_lasso_AUC <- 
  glmnet(
  x = model_matrix_X_train, y = matrix_Y_train,
  alpha = 1,
  family = "binomial",
  lambda = lambda_max_AUC_lasso
)
lambda_max_AUC_lasso_AUC

lambda_1se_AUC_lasso_AUC <- 
  glmnet(
  x = model_matrix_X_train, y = matrix_Y_train,
  alpha = 1,
  family = "binomial",
  lambda = lambda_1se_AUC_LASSO
)
lambda_1se_AUC_lasso_AUC
```

```r
coef(lambda_max_AUC_lasso_AUC)
coef(lambda_1se_AUC_lasso_AUC)
```

- If you prioritize predictive accuracy, go with max AUC.

- If you prioritize simplicity and generalizability, choose the 1se AUC lambda.

Here, I think our goal is to maximize the model's performance, so we choose the model with max AUC!

Use the model to predict the `target` variable on the **test** set. Then, use the `roc` function to obtain the ROC curve in the test set.

```r
model_matrix_X_test <- 
    model.matrix(object = PlacementStatus ~ .,
                 data = placement_testing)[, -1]

ROC_lasso <- 
    roc(
        response = placement_testing$PlacementStatus,
        predictor = predict(lambda_max_AUC_lasso_AUC,
                     newx = model_matrix_X_test)[,"s0"] ) 
ROC_lasso
```

```r
ROC_full_log_ordinary <- 
    roc(
        response = placement_testing$PlacementStatus, 
        predictor = predict(logit_model, newdata = placement_testing))
```

```r
plot(ROC_lasso, print.auc = TRUE, col = "blue", lwd = 3, lty = 2,
     main = "ROC Curves for Placement Status test set")

lines.roc(ROC_full_log_ordinary, col = "red", lwd = 3, lty = 2)
```

#### What It Means When the ROC Curves for Max Lasso AUC and the Full Logistic Model Overlap (From ChatGPT)

Both models have similar discriminatory power → They have nearly the same ability to distinguish between the positive and negative classes.

No significant predictive performance is lost when applying Lasso regularization, despite reducing the number of variables. This makes the Lasso model preferable because it is more parsimonious and less prone to overfitting.
